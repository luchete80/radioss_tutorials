```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df2_01 = pd.read_csv('force_shamrock_1tool_malla3_ms1e-5.csv')
df2_01/=10
#print (df1['f'])
#df2_01['f'].plot(label='Cone 55 Z Force', color='g')



#plt.legend(loc='upper right')
#plt.grid()


#plt.title('SPH Extrusion Bar - Reaction Force', fontsize=15)
#plt.xlabel('t [ms]')
#plt.ylabel('F[Ton]')


#window_size = 10
#df2_01['f_filtered'] = df2_01['f'].rolling(window=window_size, min_periods=1).mean()
# Define a convolution kernel (e.g., for averaging 10 values)
kernel_size = 10
kernel = np.ones(kernel_size) / kernel_size  # Averages over 10 values

# Apply the convolution operation
df2_01['f_filtered'] = np.convolve(df2_01['f'], kernel, mode='same')

average_filtered_350_to_700 = df2_01.loc[600:1000, 'f_filtered'].mean()

print(f"The average of the filtered data from step 600 to 1000 is: {average_filtered_350_to_700}")



plt.figure(figsize=(8, 5))
df2_01['f'].plot(label='Shamrock 1 Tool MS 8E-5', color='g')
#df2_01['f_filtered'].plot(label='Filtered (conv by 10)', color='r', linestyle='--')
df2_01['f_filtered'].plot(label='Filtered (conv by 10)', color='r')
plt.title('Force Data with Convolution Filter')
plt.xlabel('Index')
plt.ylabel('Force (scaled)')
plt.legend()
plt.show()
```
